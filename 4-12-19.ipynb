```
# Read a number
# Print the even digit sum from the number
# Input : 1234
# output : 6(2 + 4)
def evenDigitSum(n):
  s = 0
  while n != 0:
    r = n%10
    if r%2 == 0:
      s = s + r
    n = n // 10
  return s
evenDigitSum(1234)

```

```
# swap values
def swap(x=2,y=3):
    temp = x
    x=y
    y = temp
    return x,y
swap(2,3)
```

```
#input :19535
#output : 9(large digit of the given no)
def largeDigit(n):
    large = 0
    while n!=0:
       r=n%10
       if large < r:
           large = r
       n=n//10
    return large
largeDigit(19535)
```
